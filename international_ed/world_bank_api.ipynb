```python
import wbgapi as wb
help(wb)
```

```python
wb.source.info()
```

```python
# Get inforamtion on database 12, Education Statistics
# See full more readable list here: 
# https://datatopics.worldbank.org/education/indicators

wb.series.info(db=12)
```

```python
wb.db = 12
wb.series.info()
```

```python
wb.search('PISA')
```

```python
wb.data.DataFrame('SE.SEC.DURS')
```

```python
wb.data.DataFrame('LO.PISA.REA')
```

```python
pisa_reading_2018 = wb.data.DataFrame('LO.PISA.REA', time=2018)
pisa_reading_2018.head()
```

```python
pisa_reading_2018_clean = pisa_reading_2018.dropna()
pisa_reading_2018_clean.head()
```

```python
pisa_reading_2018_clean.shape
```

```python
pisa_math_2018 = wb.data.DataFrame('LO.PISA.MAT', time=2018)
pisa_math_2018.head()
```

```python
pisa_math_2018_clean = pisa_math_2018.dropna()
pisa_math_2018_clean.shape
```

```python
pisa_sci_2018 = wb.data.DataFrame('LO.PISA.SCI', time=2018)
pisa_sci_2018.head()
```

```python
pisa_sci_2018_clean = pisa_sci_2018.dropna()
pisa_sci_2018_clean.shape
```

```python
import pandas as pd
full_pisa = pd.concat([pisa_reading_2018_clean, pisa_math_2018_clean, pisa_sci_2018_clean], axis=1)
full_pisa.head()
```

```python
full_pisa.shape
```

```python
# Population, ages 0-14, total
student_pop_to_14 = wb.data.DataFrame('SP.POP.0014.TO', time=2018)
student_pop_to_14.head() 
```

```python
full_pisa = full_pisa.join(student_pop_to_14)
full_pisa.head()
```

```python
full_pisa.to_csv("2018_pisa_scores.csv")
```
